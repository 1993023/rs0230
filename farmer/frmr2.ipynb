```python
import pandas as pd
df = pd.read_csv("OFIS_Farmer_Name_DOB.csv") 
df1=df
df1['date'] = pd.to_datetime(df['Submitted On'])
df1['year']= df1['date'].dt.year

# df1['Module_check'] = 1
df1 = df1.drop_duplicates(subset=['Farmer Id','Module Name'],keep='last')
df2 = df1.pivot(index='Farmer Id', columns='Module Name', values='year')
# print (df2.head(2))
df2



# df2.to_csv('df2.csv')
```

```python
df5=df.pivot_table(index=['Farmer Id'], values=["Farmer Name","Gender","DOB","Product Name","Country Name","Region Name","District Name","Place Name","Farmer Group Name",
                "Partner Name","Programmer Name","Section Name","Enumerator Name","Submitted On","Survey Id","Module Name"], aggfunc='last')

# print(df5)
# df5.to_csv('df5.csv')
```

```python
df_cd = pd.merge(df5,df2, how='left', on = 'Farmer Id')
df_cd
# df_cd.to_csv('OFIS_Farmer_Name_DOB_updated.csv')
```

```python
# Another Solution
import pandas as pd
df = pd.read_csv("OFIS_Farmer_Name_DOB_Product.csv") 
```

```python
df['date'] = pd.to_datetime(df['Submitted On'])
df['year']= df['date'].dt.year

df["year"].apply(str)
df["year"]=df["year"].apply(str)
df["Module Name_updated"]=df[["Module Name","year"]].apply(lambda x:"".join(x),axis=1)
df



# df.to_csv('testt1.csv')
```

```python
# df1['Module_check'] = 1
df1=df
df1 = df1.drop_duplicates(subset=['Farmer Id','Module Name_updated'],keep='last')
df2 = df1.pivot(index='Farmer Id', columns='Module Name_updated', values='year')
# print (df2.head(2))
df2
# df2.to_csv('testt.csv')
```

```python
df3=df
df4=df3.pivot_table(index=['Farmer Id'], values=["Farmer Name","Gender","DOB","Product Name","Country Name","Region Name","District Name","Place Name","Farmer Group Name",
                "Partner Name","Programmer Name","Section Name","Enumerator Name","Submitted On","Survey Id","Module Name"], aggfunc='last')

# print(df4)
# df4.to_csv('df4.csv')
```

```python
df_cd = pd.merge(df4,df2, how='left', on = 'Farmer Id')
df_cd
df_cd.to_csv('OFIS_Farmer_Name_DOB_Product_updated.csv')
```
