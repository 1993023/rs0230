```python
import json
dlist=[]
for i in range(len(df1)):
    d = json.loads(df1["position_data"][i])
    dlist.append(d)
dlist
```

```python
infolist=[]
for j in range(len(dlist)):
    for info  in dlist[j]["location"]["plots"]:
        t=[]
        t.extend(info["plot"])
        infolist.append(t)
        break
infolist
```

```python
# dictionary into list
list1=[] 
for i in range(len(infolist)):
    t=[]
    for j in infolist[i]:
        for k, v in j.items():
            temp = [k,v]
            t.append(temp)
    list1.append(t) 
       
        
(list1[0])
```

```python
# removing ele from list
# res = [list1[i] for i in range(len(list1)) if list1[i][0] != "ele"] 
# res

res=[]
for i in range(len(list1)):
    t=[]
    for j in range(len(list1[i])):
#         print(list1[i][j])
        if list1[i][j][0] != "ele":
            t.append(list1[i][j])
    res.append(t)

res[0]
```

```python
# lat and lng in list
test1=[]
for i in range(len(res)):
    t=[]
    for j in range(len(res[i])):
#         print(res[i][j][1])
        t.append(res[i][j][1])
    test1.append(t)  
(test1[0])
```

```python
# pair of consecutive element
# test2= [[test1[i], test1[i + 1]] 
#        for i in range(len(test1) - 1)]
# test2

test2=[]
for i in range(len(test1)):
    t=[]
    for j in range(len(test1[i])-1):
#         print(test1[i][j])
        t.append([test1[i][j], test1[i][j+1]])
    test2.append(t)
test2[0]
```

```python
# removing odd element 
latlist2=[]
for i in range(len(test2)):
    t=[]
    for j in range(len(test2[i])):
        if j%2!=0:
            continue
        else:
            t.append(test2[i][j])
    latlist1=[]
    latlist1.append(t)
    latlist2.append(latlist1)

# latlist
nwlatlist=[]
nwlatlist.append(latlist2)
# (latlist)
(nwlatlist[0])
```
