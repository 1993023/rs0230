```python
import json
d = json.loads(df1["position_data"][0])
d
```

```python
infolist=[]
for info  in d["location"]["plots"]:
    infolist.extend(info["plot"])
infolist
```

```python
# dictionary into list
list1=[]
for i in range(len(infolist)):
    for k, v in infolist[i].items():
        temp = [k,v]
        list1.append(temp)          
list1

# removing ele form list
res = [list1[i] for i in range(len(list1)) if list1[i][0] != "ele"] 
res

# lat and lng in list
test1=[]
for i in res:
    test1.append(i[1])  
test1  

# pair of consecutive element
test2= [[test1[i], test1[i + 1]] 
       for i in range(len(test1) - 1)]
test2

# removing odd element 
latlist=[]
for i in range(0,len(test2)):
    if i%2!=0:
        continue
    else:
        latlist.append(test2[i])

latlist
```

```python
import pandas as pd
import geojson

def data2geojson(df):
    features = []
    insert_features = lambda X: features.append(
            geojson.Feature(geometry=geojson.Polygon((X["lat"],
                                                    X["long"]))))
    df.apply(insert_features, axis=1)
    with open('test1.geojson', 'w', encoding='utf8') as fp:
        geojson.dump(geojson.FeatureCollection(features), fp, sort_keys=True, ensure_ascii=False)

col = ['lat','long']
data = latlist

test_df = pd.DataFrame(data, columns=col)

data2geojson(test_df)
```

```python
# geojson to shapefile
import urllib, geojson, gdal, subprocess
with open('test1.geojson') as f:
    data = geojson.load(f)

with open('test2.geojson', 'w') as f:
    geojson.dump(data, f)

args = ['ogr2ogr', '-f', 'ESRI Shapefile', 'test2.shp', 'test2.geojson']
subprocess.Popen(args)
```
